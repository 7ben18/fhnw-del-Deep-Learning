```python
# Module importieren
import torch
import numpy as np
```

# Creation 


```python
# Torch erstellen (matrizen)  
display("Create empty torch", torch.empty(5, 3))
display("Create zeros torch", torch.zeros(2, 3))
display("Create torch with values", torch.full((2, 3), 3.14))
display("Create torch sequence", torch.tensor((10,5,10)))
display("Create eye torch", torch.eye(3))
display("Create torch from numpy", torch.from_numpy(np.array([[1,2,3],[4,5,6]])))


# Torch erstellen (zuf√§llig)
tensor = torch.rand(3, 3)
display("Create random torch", tensor)
```

```python
# Tensor anschauen (tensor, zeilen, spalten)
display(torch.Tensor.view(tensor, 3, 3))
display(torch.Tensor.view(tensor, 1, 9))

# Tensor concardieren (tensor, tensor, dimension)
display("Horizontal concat", torch.cat((tensor, tensor), 0))
display("Vertical concat", torch.cat((tensor, tensor), 1))

# Tensor unterteilen in Zeilen oder Spalten (tensor, anzahl, dimension)
display("Tensor in chunks unterteilen", torch.chunk(tensor, 3, 0))
display("Tensor in chunks unterteilen", torch.chunk(tensor, 3, 1))
display("Tensor in chunks unterteilen", torch.chunk(tensor, 2, 0))
```
